```python

def get_ticker_prices(excel_file_path,tab_name,frequency):
    #Interval options: 1d, 5d, 1w, 1m, 3m
    
    #GET TICKER LISTS
    ticker_df = pd.read_excel(excel_file_path, sheet_name=tab_name)
    
    global tickers
    global tickers_and_names

    tickers= ticker_df['Symbol']
    tickers_and_names = ticker_df
    
    dfs = []

    for ticker in tickers:
        try:
            data = yf.download(ticker, start=start_date, end=end_date, interval = frequency, auto_adjust=False)
            data[f'{ticker} Open'] = data['Open']
            data[f'{ticker} High'] = data['High']        
            data[f'{ticker} Low'] = data['Low']        
            data[f'{ticker} Close'] = data['Close']
            data[f'{ticker} Adj Close'] = data['Adj Close']            
            data[f'{ticker} Volume'] = data['Volume']
            
            data = data[[f'{ticker} Open', f'{ticker} High', f'{ticker} Low', f'{ticker} Close', f'{ticker} Volume', f'{ticker} Adj Close']]
            dfs.append(data)
        except:
            pass

    df = pd.concat(dfs, axis=1)
    df.index = pd.to_datetime(df.index)
    df.columns = df.columns.get_level_values(0) #removes the multilevel index
        
    return df

```

```python
#Choose market (can test on 'Short ticker list')
market = 'ETF universe slim 3'

#Access tickers in file
excel_file = r'C:\EXCEL PATH TO FILE WITH LIST OF TICKERS\ETF list.xlsx'
excel_export_path = r'C:\EXCEL PATH TO SAVE PRICE HISTORIES\Temporary price storage file.xlsx'

# Get the NYSE trading calendar
nyse = get_calendar('XNYS')

# Specify the end date
calendar_start_date = '1997-01-01'
calendar_end_date = '2025-05-12'
frequency = '1wk' #1d, 1wk, 1mo, 3mo
frequency_multiply_factor = 7 #1:day, 7: week, 30: month, 63:quarter
periods_to_download = 13 #Number of periods (days, weeks, months, etc) to download
data_download_history = periods_to_download*frequency_multiply_factor 
sheet_name = 'Weekly'

OHLC = 'Close'
trend_numerator = 1
trend_denominator = 10
number_of_positions = 10

spread_cost = .1 #MEASURE IN PERCENT

end_date = datetime.datetime.strptime(calendar_end_date, "%Y-%m-%d").date() + datetime.timedelta(days=1) #USED FOR get_ticker_prices
start_date = datetime.datetime.strptime(calendar_start_date, "%Y-%m-%d").date()

#pull price histories as of each day #PULLS AND STORES OPEN, HIGH, LOW, CLOSE, VOLUME
start_date = start_date - datetime.timedelta(days=data_download_history) #change start_date to get necessary history

df = get_ticker_prices(excel_file,market,frequency)

# Write df to the 'Daily' sheet without overwriting other sheets
with pd.ExcelWriter(excel_export_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    df.to_excel(writer, sheet_name=sheet_name, index=True)
```
